```python
eia_api_key = "YOUR KEY"

url = "http://api.eia.gov/series/"
series_id = "NG.N9100US3.M"


query = f"{url}?api_key={eia_api_key}&series_id={series_id}"
response = requests.get(query).json()
#pprint(response)

data = response["series"][0]["data"]
month = []
val = []
for x in data:
    date_str = x[0][4:]+"/15/"+x[0][:4]
    date_object = datetime.strptime(date_str,'%m/%d/%Y').date()
    #print(f"{(date_object)} , {float(x[1])}")
    month.append(date_object)    
    val.append(float(x[1]))

title = response["series"][0]["name"]
country = response["series"][0]["geography"]
unit = response["series"][0]["units"]



large = 18; med = 14; small = 10
params = {'axes.titlesize': med,
          'legend.fontsize': med,
          'figure.figsize': (15, 5),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")

fig, ax = plt.subplots()
fig.suptitle(title, fontweight="bold")
ax.set_xlim(min(month), max(month))
ax.set_ylim(0, max(val) + 2)

#ax.set_xlabel("Fake Banana Ages (in days)")
ax.set_ylabel(unit, rotation=0, labelpad=20)
ax.yaxis.set_label_coords(0.09,1.03)
ax.yaxis.grid(True)

ax.plot(month,val, linewidth=2,color="Slateblue")
plt.show()

```

```python
def ExtractData(url,eia_api_key,series_id):
    query = f"{url}?api_key={eia_api_key}&series_id={series_id}"
    response = requests.get(query).json()
    #pprint(response)

    data = response["series"][0]["data"]
    valx = []
    valy = []
    for x in data:
        date_str = x[0][4:]+"/15/"+x[0][:4]
        date_object = datetime.strptime(date_str,'%m/%d/%Y').date()
        #print(f"{(date_object)} , {float(x[1])}")
        valx.append(date_object)    
        valy.append(float(x[1]))

    title = response["series"][0]["name"]
    country = response["series"][0]["geography"]
    unit = response["series"][0]["units"]
    
    return valx, valy, title, country, unit


def PlotData(valx, valy, title, country, unit):
    large = 18; med = 14; small = 10
    params = {'axes.titlesize': med,
              'legend.fontsize': med,
              'figure.figsize': (15, 5),
              'axes.labelsize': med,
              'axes.titlesize': med,
              'xtick.labelsize': med,
              'ytick.labelsize': med,
              'figure.titlesize': large}
    plt.rcParams.update(params)
    plt.style.use('seaborn-whitegrid')
    sns.set_style("white")

    fig, ax = plt.subplots()
    fig.suptitle(title, fontweight="bold")
    ax.set_xlim(min(valx), max(valx))
    ax.set_ylim(0, max(valy) + 2)

    #ax.set_xlabel("Fake Banana Ages (in days)")
    ax.set_ylabel(unit, rotation=0, labelpad=20)
    ax.yaxis.set_label_coords(0.09,1.03)
    ax.yaxis.grid(True)

    ax.plot(valx,valy, linewidth=2,color="Slateblue")
    plt.show()
    
    return fig
```

```python
eia_api_key = "YOUR KEY"
url = "http://api.eia.gov/series/"
series_id = "NG.N9100US3.M"

valx, valy, title, country, unit = ExtractData(url,eia_api_key,series_id)
x = PlotData(valx, valy, title, country, unit)
```

```python
url = "https://api.worldweatheronline.com/premium/v1/past-weather.ashx"
api_key = "YOUR KEY"
city = "New York, NY"
#q=New+York
#q=New+york,ny
#q=London,united+kingdom
ini_date = "2012-07-1"
end_date = "2012-07-31"
out_format = "json"
NumHours = 24

params = {
    "q": city,
    "date": ini_date,
    "enddate": end_date,
    "key": api_key,
    "format": out_format,
    "tp": NumHours
}

response = requests.get(url, params=params)
data = response.json()
pprint(data)
```

```python
x = data["data"]["weather"]

valx = [] 
valy = []
for i in range(0,len(x),1):
    print(f' {x[i]["date"]} : {x[i]["maxtempC"]} C ')
    date_object = datetime.strptime(x[i]["date"],'%Y-%m-%d').date()
    valx.append(date_object)
    #valx.append(x[i]["date"])
    valy.append(float(x[i]["maxtempC"]))
```

```python
title = "Montly Maximum Temperature"
country = "US"
unit = "C"
x = PlotData(valx, valy, title, country, unit)
```
